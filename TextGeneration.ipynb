```python
import markovify
```

```python
with open("data/combined.txt") as f:
    text = f.read()

# Build the model.
text_model = markovify.Text(text, well_formed=False)

# Print five randomly-generated sentences
for i in range(5):
    print(text_model.make_sentence())
```

```python
for i in range(100):
    print(text_model.make_short_sentence(100000) + '\n')
```
